```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def get_word_fruency(words):
    wordfreq = {}
    for word in words:
        if word not in wordfreq:
            wordfreq[word] = 0
        wordfreq[word] += 1
    return wordfreq
```

```python
df = pd.read_csv("../data/news.csv")
text_col = df["title"]

text_pool = ""
for a in df["title"]:
    text_pool += a

_char_to_replace = [",", "'", '"', ".", ":", ";", "!", "?", "|", "&"]
for char in _char_to_replace:
    text_pool = text_pool.replace(",", "")
word_pool = text_pool.split()

dic = get_word_fruency(word_pool)
dic_cleaned = {}
for key in dic.keys():
    if (dic[key] > 50) and (dic[key] % 5 ==0):
        dic_cleaned[key] = dic[key]

print(dic_cleaned)
```

```python
numbers = dic_cleaned.values()
labels = dic_cleaned.keys()

explode = (0.1, 0, 0, 0)
fig1, ax1 = plt.subplots()
ax1.pie(numbers,  labels=labels,
        shadow=True, startangle=90,
        autopct='%1.1f%%')
ax1.axis('equal')
plt.show()
```
